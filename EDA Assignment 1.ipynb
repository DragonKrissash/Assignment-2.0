```python
Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in
    predicting the quality of wine.
```

```python
A1. The key features of the wine quality dataset are fixed acidity, volatile acidity, citric acid, residual
    sugar, chlorides, free sulfur dioxide, density, pH, sulphates, alcohol and quality.
    
    Fixed acidity - Refers to total concentration of acids in wine. It affects overall taste and stability of
    wine. Higher acidity refers to more refreshing and lively taste.
    
    Volatile acidity - It refers to presence of volatile acids like acetic acid which contributes to vinegar
    like smell and taste. Higher the value, impacts the wine quality negatively.
    
    Citric acid - It is a weak organic acid that influences wine's tartness and freshness. It can enhance the 
    wine's overall balance.
    
    Residual Sugar - Residual sugar is the amount of sugar remaining in the wine after fermentation. It 
    contributes to the wine's sweetness and can balance out acidity, affecting the perceived taste and 
    overall harmony of flavors.
    
    Chlorides - Chlorides represent the concentration of salts in the wine. They can contribute to the 
    wine's taste, mouthfeel, and perceived bitterness. Proper balance is crucial for a pleasant tasting 
    experience.

    Free Sulfur Dioxide - Free sulfur dioxide acts as a preservative in wine, preventing spoilage and 
    oxidation. It can also influence the wine's aroma and flavor.
    
    Total Sulfur Dioxide - Total sulfur dioxide includes both free and bound forms of sulfur dioxide. 
    It's an important indicator of wine stability and can affect the overall quality and aging potential 
    of the wine.

    Density - Density is a measure of the wine's mass relative to its volume. It can provide insights into 
    the wine's concentration and level of residual sugar.

    pH - pH measures the acidity or alkalinity of the wine. It significantly influences the wine's taste, 
    stability, and overall balance. Wines with appropriate pH levels tend to have better aging potential.

    Sulphates - Sulphates, or sulfites, are additives that act as antioxidants and antimicrobial agents in 
    wine. They can affect the wine's aroma, flavor, and overall quality.

    Alcohol - Alcohol content affects the wine's body, mouthfeel, and perceived warmth. It can contribute to 
    the wine's overall balance and harmony of flavors.

    Quality - Quality is the target variable, and it represents the subjective rating of the wine's quality 
    by human tasters. It's typically scaled on a scale of 1 to 10, where higher values indicate better quality.
```

```python
Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
Discuss the advantages and disadvantages of different imputation techniques.
```

```python
A2. There are no missing data in wine quality dataset to handle. We can handle the missing data through 
    various processes like:
    
    Mean/Median imputation - This is replacing the missing data with the average of the other data. This is 
    helpful only if the missing data is completely at random and can lead to biased estimates if the missing
    data is not random.
    
    Mode imputation - It is mostly suited for categorical data where the missing data is replaced with the 
    most frequently occuring data.
    
    Forward fill/ Backward fill - This technique uses the last observed value for forward fill and next 
    observed value for backward fill. This is only useful when there is logical continuity in the data.
```

```python
Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
features. What is the minimum number of principal components required to explain 90% of the variance in
the data?
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
df=pd.read_csv('winequality-red.csv')
x=df.drop('quality',axis=1)
scaler=StandardScaler()
x_sc=scaler.fit_transform(x)
pca=PCA()
x_pca=pca.fit_transform(x_sc)
exp_var=pca.explained_variance_ratio_
cum_exp_var=exp_var.cumsum()
num_com=(cum_exp_var>=0.9).argmax()+1
print(f"Minimum number of components to explain 90% variance: {num_com}")
```

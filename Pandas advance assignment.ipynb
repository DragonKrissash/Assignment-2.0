```python
Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
'Word_Count' that contains the number of words in each row of the 'Text' column.
```

```python
def word_count(x):
    return len(x.split())
df=pd.DataFrame({'Text':['My name is Krishna','I am going to enter college','I have taken data science']})
```

```python
df
```

```python
df['Word_Count']=df['Text'].apply(word_count)
```

```python
df
```

```python
Q5. How are DataFrame.size() and DataFrame.shape() different?
A5. DataFrame.size() returns total number of elements in dataframe by multiplying rows and columns while
    DataFrame.shape() returns a tuple which give the number of rows and columns in the form (rows,columns).
```

```python
Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.
```

```python
df=pd.DataFrame({'Email':['krish@gmail.com','sudh@gmail.com','kumar@gmail.com']})
```

```python
df
```

```python
l=[]
for i in df['Email']:
    ind=i.index('@')
    l.append(i[0:ind])
l
```

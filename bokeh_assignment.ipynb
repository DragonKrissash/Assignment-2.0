Q1. How can you create a Bokeh plot using Python code?

A1. First we need to install the bokeh plot. Then we need to import modules like bokeh.plotting, bokeh.io.
    Then we need to prepare the data to be plotted and create a figure to plot the data and show it.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show
import numpy as np
x=np.random.rand(10)
y=np.random.rand(10)
p=figure()
p.circle(x, y, size=10, color='navy', alpha=0.5)
p.xaxis.axis_label = 'X-axis'
p.yaxis.axis_label = 'Y-axis'
p.title.text = 'Bokeh Plot'
p.background_fill_color = 'lightgray'
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

A2. In Bokeh glyphs are visual markers or symbols that represent data on plot. They can be used to create various types of plots like scatter plot, line plot, barchart.
To add glyphs to Bokeh plot we need to use the figure method.

```python
from bokeh.plotting import figure,show
p=figure()
p.circle(x=[1,2,3,4,5],y=[2,5,8,3,6], size=10, color='navy', alpha=0.5)
p.line(x=[1, 2, 3, 4, 5], y=[2, 5, 8, 3, 6], line_width=2, color='red')
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
#axes can be customized using the following functions
p.xaxis.axis_label = "X-axis Label"  # Set the label for the x-axis
p.yaxis.axis_label = "Y-axis Label"  # Set the label for the y-axis
p.xaxis.axis_label_text_color = "red"  # Set the color of the x-axis label
p.yaxis.major_label_text_font_size = "12pt"  # Set the font size of the y-axis major labels
p.xaxis.major_tick_line_color = None  # Hide the major tick lines on the x-axis
p.yaxis.minor_tick_line_color = "gray" # Set the color of the minor tick lines on the y-axis
```

```python
#title can be customized using the following functions
p.title.text = "Custom Plot Title"  # Set the title text
p.title.text_font_size = "16pt"  # Set the font size of the title
p.title.text_color = "blue"  # Set the color of the title
p.title.align = "center"  # Align the title to the center
```

```python
#Legend can be customized using the following functions
p.legend.location = "top_left"  # Set the location of the legend (e.g., top-left, top-right)
p.legend.title = "Legend"  # Set the title of the legend
p.legend.label_text_font_style = "italic"  # Set the font style of the legend labels
p.legend.background_fill_alpha = 0.5  # Set the transparency of the legend background
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

A5. TO embed a Bokeh plot into a web page or dashboard using flask or Django we need to follow following path. Below one is for flask

```python
pip install flask
pip install bokeh

from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create the Bokeh plot
    p = figure()
    p.circle([1, 2, 3], [4, 5, 6])

    # Generate the JavaScript and HTML components for embedding
    script, div = components(p)

    # Render the template with the plot components
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)
```

```python
<!DOCTYPE html>
<html>
<head>
    <title>Bokeh Plot</title>
    {{ script|safe }}  <!-- Embed the Bokeh plot JavaScript -->
</head>
<body>
    <h1>Bokeh Plot</h1>
    <div>
        {{ div|safe }}  <!-- Embed the Bokeh plot HTML -->
    </div>
</body>
</html>
```

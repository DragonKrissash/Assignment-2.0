```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

```python
A2. Scatter plot is a type of data visualization where individual data points are marked on a 2D plane. It is
    commonly used to show relationship or correlation between 2 variables.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Random number plot')
plt.plot(x,y)
```

```python
Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3,4,5])
y1=np.array([0, 100, 200, 300, 400, 500])
y2=np.array([50, 20, 40, 20, 60, 70])
y3=np.array([10, 20, 30, 40, 50, 60])
y4=np.array([200, 350, 250, 550, 450, 150])
fig,axs=plt.subplots(2,2)
axs[0,0].plot(x,y1)
axs[0, 0].set_title('Line Plot 1')
axs[0,1].plot(x,y2)
axs[0, 1].set_title('Line Plot 2')
axs[1,0].plot(x,y3)
axs[1, 0].set_title('Line Plot 3')
axs[1,1].plot(x,y4)
axs[1,1].set_title('Line Plot 4')
plt.tight_layout()
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
A4. A bar plot or bar chart is a type of data visualization that uses rectangular bars to represent 
    categorical data. It is used to display frequencies, counts or values of different categories.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
plt.title('Profit of companies')
plt.xlabel('Companies')
plt.ylabel('Profit')
plt.bar(company,profit,color=['r','b','g','y'])
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
A5. A box plot is also known as box and whisker plot which is a type of statistical data visualisation that 
    display distribution of dataset through max, min, 1st quartile, median, 2nd quartile.
    Box provides visual representation of data distribution, particularly skewness, symmetry, and presence of
    outliers.
```

```python
import numpy as np
import matplotlib.pyplot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
labels=['Box 1','Box 2']
plt.boxplot(data,labels=labels)
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
